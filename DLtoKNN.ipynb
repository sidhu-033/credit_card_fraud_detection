```python
from tensorflow.keras.models import load_model
import pickle
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv('creditcard.csv')
scaler = StandardScaler()
df['NormalizedAmount'] = scaler.fit_transform(df['Amount'].values.reshape(-1, 1))
df = df.drop(['Amount', 'Time'], axis = 1)
```

```python
df3=pd.merge(df1,df2,how='outer')
```

```python
final_df=df3.sample(frac=1)
```

```python
final_df
```

```python
#load model
model = load_model('model02.h5')
filename='Iforest_model.sav'
Iforest = pickle.load(open(filename, 'rb'))
model1 = load_model('model01.h5')
filename1 = 'knn_model.sav'
knn = pickle.load(open(filename1, 'rb'))
```

```python
y = final_df['Class']
X = final_df.drop(['Class'], axis = 1)
```

```python
def predict(X):
    y=[]
    for i in range(len(X)):
        print(i)
        x=np.array(X.iloc[i])
        x=x.reshape(1,29)
        prediction=model.predict_classes(x, verbose=0)
        if prediction==1:
            print("making prediction using knn")
            prediction2=knn.predict(x)
            #prediction3=model1.predict_classes(x, verbose=0)
            if prediction==1:
                y.append(1)
            else:
                y.append(0)
        else:
            y.append(0)
    return y
```

```python
predicted_classes1 = predict(X)
```

```python
predictions1=np.array(predicted_classes1)
predictions1.shape
predictions1=predictions1.reshape(984,1)
predictions1.shape
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import f1_score
from sklearn.metrics import roc_auc_score
from sklearn.metrics import confusion_matrix
```

```python
# accuracy: (tp + tn) / (p + n)
accuracy = accuracy_score(y, predictions1)
print('Accuracy: %f' % accuracy)
# precision tp / (tp + fp)
precision = precision_score(y, predictions1)
print('Precision: %f' % precision)
# recall: tp / (tp + fn)
recall = recall_score(y, predictions1)
print('Recall: %f' % recall)
# f1: 2 tp / (2 tp + fp + fn)
f1 = f1_score(y, predictions1)
print('F1 score: %f' % f1)
# confusion matrix
matrix = confusion_matrix(y, predictions1)
print(matrix)
```
